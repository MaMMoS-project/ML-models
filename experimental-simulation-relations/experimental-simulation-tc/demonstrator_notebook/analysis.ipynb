# Error Correction Curie temperature

This notebook showcases how to train machine learning models for error correction from simulation results to experimental measurements. As inputs the models take the simulated value of a magnetic material, optionally along with a compound embedding of the stoichiometric formula. The material data is analyzed in three groups: all materials, rare-earth materials and rare-earth free materials. In order to increase the number of training samples, data augmentation is performed. Here missing experimental and simulated values are generated.

### Dataset 

The dataset is compiled from different sources of literature and exisiting databases. That is data from spin-dynamics simulations, data from publications and data of magnetic materials published in the NEMAD database. From these data pools we were able to collect in total 839 pairs: from these pairs 547 samples are for “rare-earth free” (RE-free) materials and 292 samples for “rare-earth” (RE) materials.
